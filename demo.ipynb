```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.imdb.com/chart/top/?ref_=chtstrm_ql_3'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

movies = []
ratings = []
timings = []
years = []
vote_counts = []

allowed_ratings = ['R', 'PG', 'PG-13', 'Not Rated', 'Approved','A','U','UG','NA','UA']

# Locate all movie blocks
uls = soup.find_all('li', class_='ipc-metadata-list-summary-item')

for li in uls:
    # Movie title
    title_tag = li.find('h3')
    if title_tag:
        movies.append(title_tag.text.strip().split('. ', 1)[1])
    
    # Rating
    rating_tag = li.find('span', class_="ipc-rating-star--rating")
    if rating_tag:
        ratings.append(rating_tag.text.strip())
    else:
        ratings.append(None)

    # Vote count
    vote_tag = li.find('span', class_='ipc-rating-star--voteCount')
    if vote_tag:
        clean_vote = vote_tag.text.replace('(', '').replace(')', '').strip()
        vote_counts.append(clean_vote)
    else:
        vote_counts.append(None)

    # Metadata: year, runtime, etc.
    meta_tags = li.find_all('span', class_="sc-15ac7568-7 cCsint cli-title-metadata-item")
    year_val = None
    for span in meta_tags:
        text = span.text.strip()
        if text.isdigit():
            year_val = text
        elif text in allowed_ratings:
            continue
        else:
            timings.append(text)
    years.append(year_val)

# Output results
print((movies))
print((ratings))
print((vote_counts))
print((years))
print((timings))
```

```python
import pandas as pd
df=pd.DataFrame({'Movie':movies,'Rating':ratings,'Votes':vote_counts,'Year':years,'Duration':timings})
df.reset_index(drop=True, inplace=True)
df
```
